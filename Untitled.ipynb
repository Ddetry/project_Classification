```python
import pandas as pd
import numpy as np
import matplotlib
import os
from sklearn import tree
from sklearn import preprocessing
%matplotlib inline

path = '/Users/damien/Documents/Sinf/2eme_bis/2eme_quadri/QDA/project_Classification/BDD-performance.csv'

my_df = pd.read_csv(path, sep=';')

label_encoder = preprocessing.LabelEncoder()

# Convert Sex variable to numeric
encoded_sex = label_encoder.fit_transform(my_df["sex"])

# Initialize model
tree_model = tree.DecisionTreeClassifier()

# Train the model
tree_model.fit(X = pd.DataFrame(encoded_sex), 
               y = my_df["age"])

with open("tree1.dot", 'w') as f:
     f = tree.export_graphviz(tree_model, 
                              feature_names=["sex"], 
                              out_file=f)
```

```python
        from IPython.display import Image
Image("tree1.png") 

preds = tree_model.predict_proba(X = pd.DataFrame(encoded_sex))

pd.crosstab(preds[:,0], my_df["sex"])
```

```python
# Make data frame of predictors
predictors = pd.DataFrame([encoded_sex, my_df["traveltime"]]).T

# Train the model
tree_model.fit(X = predictors, 
               y = my_df["sex"])
```

```python
with open("tree2.dot", 'w') as f:
     f = tree.export_graphviz(tree_model, 
                              feature_names=["Sex", "Class"], 
                              out_file=f)
```

```python
# Get survival probability
preds = tree_model.predict_proba(X = predictors)

# Create a table of predictions by sex and class
pd.crosstab(preds[:,0], columns = [my_df["paid"], my_df["romantic"],my_df["age"], my_df["higher"], my_df["schoolsup"],my_df["failures"], my_df["goout"]])

```
